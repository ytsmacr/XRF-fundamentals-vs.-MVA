```python
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.collections import PatchCollection
from matplotlib.patches import Rectangle
from matplotlib.colors import to_rgba

fp='H:\\My Drive\\PROJECTS\\PSI 2022-2025\\XRF fundamentals vs. MVA'
```

```python
# inferred from samples AND lowest non-zero concentration from standards
meta = pd.read_csv(f'{fp}\\data\\meta_filter1_stratified_outliers_removed.csv')
pred = pd.read_csv(f'{fp}\\data\\instrument_predictions.csv')
elem_list = [x.split()[0] for x in pred.columns if '+/-' in x]

ilod_df = {}

for elem in elem_list:
    col = [x for x in pred.columns if (elem in x) and ('+/-' not in x)][0]
    m_p = min(pred[col].dropna())
    m_c = min(meta[meta[elem]>0][elem])
    
    ilod_df[elem]=[m_p,m_c]
    
ilod_df = pd.DataFrame.from_dict(ilod_df, orient='index').reset_index()
ilod_df.columns = ['element','min_pred','min_sample']
```

```python
# from Olympus LOD sheet
df = pd.read_csv(f'{fp}\\olympus_LODs.csv')

# convert elements to oxides
ox = ['SiO2','TiO2','Al2O3','Fe2O3','MgO','MnO','CaO','P2O5','SO3']
conv = pd.read_csv(fp+'\\misc\\element_oxide_conversion.csv')
ox_el_key = dict(zip(conv.oxide, conv.element))
el_conv_key = dict(zip(conv.element, conv.conversion))

to_add = []
for m in ox:
    elem = ox_el_key[m]
    factor = el_conv_key[elem]
    
    data = df.loc[df.element==elem].values[0][1:]
    # times factor
    data = data * factor
    # to wt%
    data = data / 10000
    data = [m]+list(data)
    
    to_add.append(data)
    
df = pd.concat([df, pd.DataFrame(to_add, columns=df.columns)], ignore_index=True)
df = df[df.element.isin(elem_list)].reset_index(drop=True)

# add inferred LODs
df = df.merge(ilod_df, how='left').sort_values('element', ignore_index=True)
df.to_csv(fp+'\\Olympus_LOD_summary.csv', index=False)
```

```python
major = df[df.element.isin(ox)].reset_index(drop=True)
# sort based on my choice
major['elem_cat'] = pd.Categorical(
    major['element'], 
    categories=ox, 
    ordered=True
)
major = major.sort_values('elem_cat', ignore_index=True)

minor = df[~df.element.isin(ox)].reset_index(drop=True)
```

```python
#make rectangles
rcolor = '#8b8b8b'
c='black'
p='darkred'
alpha=0.5

# major
box_list = []
for i in major.index:
    x = i-0.25
    y = major.iloc[i]['lower']
    width = 0.5
    height = major.iloc[i]['diff']
    
    box_list.append(Rectangle((x,y), width, height))
    
maj_pc = PatchCollection(box_list, edgecolor = to_rgba(rcolor, alpha), facecolor=to_rgba(rcolor, alpha))

# minor
box_list = []
for i in minor.index:
    x = i-0.25
    y = minor.iloc[i]['lower']
    width = 0.5
    height = minor.iloc[i]['diff']
    
    box_list.append(Rectangle((x,y), width, height))
    
min_pc = PatchCollection(box_list, edgecolor = to_rgba(rcolor, alpha), facecolor=to_rgba(rcolor, alpha))

# PLOT
fig, ax = plt.subplots(nrows=2, figsize = (6,5))

# major
ax[0].scatter(major['element'], major['min_pred'], marker='_', s=150, c=c, label = 'Minimum instrument prediction')
ax[0].scatter(major['element'], major['min_sample'], marker='x', s=40, c=p, label = 'Minimum non-zero standard concentration')
ax[0].add_collection(maj_pc)
ax[0].legend()

# minor
ax[1].scatter(minor['element'], minor['min_pred'], s=75, marker='_', c=c)
ax[1].scatter(minor['element'], minor['min_sample'], s=20,  marker='x', c=p)
ax[1].add_collection(min_pc)
ax[1].set_ylim((-1,20))

plt.ylabel('Concentration')
plt.tight_layout()
plt.savefig(fp+'\\figures\\Olympus_LOD_comparison.eps', dpi=600)
plt.close()
```
